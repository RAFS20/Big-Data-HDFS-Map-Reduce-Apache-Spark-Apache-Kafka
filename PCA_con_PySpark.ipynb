# Análisis de Componentes Principales (PCA)

El Análisis de Componentes Principales (PCA, por sus siglas en inglés) es una técnica utilizada en estadísticas y aprendizaje automático para la reducción de la dimensionalidad de datos. Esta técnica permite transformar un conjunto de datos original con muchas variables en un conjunto de datos nuevo con un número menor de variables, denominadas componentes principales. El PCA busca encontrar una proyección lineal de los datos originales en un espacio de menor dimensión de tal manera que se maximice la varianza de los datos proyectados.

### Objetivo del PCA
El objetivo principal del PCA es reducir la dimensionalidad de los datos, manteniendo la mayor cantidad de información posible. Esto se logra mediante la transformación de los datos originales en un nuevo conjunto de variables llamadas componentes principales. Para comprender el PCA desde un punto de vista matemático, primero debemos definir algunas notaciones clave:

- Sea $X$ una matriz de datos $n \times p$, donde $n$ es el número de observaciones y $p$ es el número de variables.

### Paso 1: Estandarización de datos
El primer paso en el PCA es estandarizar los datos, lo que significa que cada variable se centra en cero y tiene una desviación estándar de uno. Esto se hace para que todas las variables tengan la misma escala y no se vean afectadas por la magnitud de las unidades de medida. La estandarización se realiza de la siguiente manera:

Para cada variable $X_j$, calculamos su media $\mu_j$ y su desviación estándar $\sigma_j$ en el conjunto de datos $X$. Luego, estandarizamos $X_j$ como sigue:

$$
Z_j = \frac{X_j - \mu_j}{\sigma_j}
$$

Donde:
- $Z_j$ es la variable estandarizada.
- $X_j$ es la variable original.
- $\mu_j$ es la media de $X_j$.
- $\sigma_j$ es la desviación estándar de $X_j$.

### Paso 2: Matriz de Covarianza
Una vez que los datos se han estandarizado, calculamos la matriz de covarianza de $Z$. La matriz de covarianza $C$ es una matriz simétrica que mide las relaciones lineales entre las variables estandarizadas. Su fórmula es la siguiente:

$$
C = \frac{1}{n-1} Z^T Z
$$

Donde:
- $C$ es la matriz de covarianza.
- $Z$ es la matriz de datos estandarizados.
- $n$ es el número de observaciones.

### Paso 3: Cálculo de Autovalores y Autovectores
El siguiente paso es calcular los autovalores y autovectores de la matriz de covarianza $C$. Los autovalores representan la varianza explicada por cada componente principal, y los autovectores son las direcciones en las que se encuentra la mayor variabilidad de los datos. Los autovalores y autovectores se obtienen resolviendo la siguiente ecuación:

$$
C \mathbf{v} = \lambda \mathbf{v}
$$

Donde:
- $\mathbf{v}$ es un autovector.
- $\lambda$ es un autovalor.

El objetivo es encontrar todos los autovalores $\lambda_i$ y sus correspondientes autovectores $\mathbf{v}_i$. Estos están ordenados de mayor a menor importancia en términos de la varianza que explican.

### Paso 4: Selección de Componentes Principales
Una vez que hemos calculado los autovalores y autovectores, seleccionamos las componentes principales que queremos conservar. Esto se hace ordenando los autovalores de mayor a menor y eligiendo las $k$ componentes principales más importantes, donde $k$ es la cantidad de varianza que deseamos retener. Por lo general, se elige un valor $k$ que explica un porcentaje suficientemente alto de la varianza total de los datos, como el 95% o el 99%.

### Paso 5: Proyección de Datos en las Componentes Principales
Una vez seleccionadas las componentes principales, proyectamos los datos originales en el espacio definido por estas componentes. La proyección de un punto $x$ en las componentes principales se realiza de la siguiente manera:

$$
\text{Proyección}(x) = x \cdot V_k
$$

Donde:
- $x$ es un punto de datos.
- $V_k$ es una matriz formada por las $k$ columnas principales de la matriz de autovectores $V$.

### Paso 6: Varianza Explicada
Es importante evaluar cuánta varianza de los datos originales se conserva en las componentes principales seleccionadas. La varianza explicada por las $k$ componentes principales se calcula como la suma de los $k$ autovalores correspondientes:

$$
\text{Varianza Explicada} = \sum_{i=1}^{k} \lambda_i
$$

Este valor se utiliza para determinar cuánta información se ha conservado en la reducción de dimensionalidad.

### Resumen de las Fórmulas del PCA
En resumen, las fórmulas clave para el PCA son las siguientes:

1. Estandarización de datos:
   $$
   Z_j = \frac{X_j - \mu_j}{\sigma_j}
   $$

2. Matriz de Covarianza:
   $$
   C = \frac{1}{n-1} Z^T Z
   $$

3. Cálculo de Autovalores y Autovectores:
   $$
   C \mathbf{v} = \lambda \mathbf{v}
   $$

4. Selección de Componentes Principales:
   - Ordenar los autovalores de mayor a menor.
   - Elegir las $k$ componentes principales más importantes.

5. Proyección de Datos en las Componentes Principales:
   $$
   \text{Proyección}(x) = x \cdot V_k
   $$

6. Varianza Explicada:
   $$
   \text{Varianza Explicada} = \sum_{i=1}^{k} \lambda_i
   $$

El PCA es una técnica fundamental en estadísticas y aprendizaje automático que permite la reducción de la dimensionalidad de los datos mientras se conserva la mayor cantidad posible de información.

```python
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withStd=True, withMean=False)
```

En este paso, se crean dos etapas dentro de un pipeline. La primera etapa, `assembler`, se utiliza para ensamblar las características en un solo vector, que es una entrada requerida para PCA.
```python
# Paso 2: Aplicar PCA
pca_initial = PCA(inputCol="scaledFeatures", outputCol="pcaFeatures", k=len(X_spark.columns))
pipeline = Pipeline(stages=[assembler, scaler, pca_initial])
model = pipeline.fit(X_spark)
transformed = model.transform(X_spark)
```

En este paso, se realiza la primera ejecución de PCA. Se crea una instancia de la clase `PCA` con la entrada estandarizada y se especifica el número de componentes `k` igual al número de columnas en el DataFrame `X_spark`. Luego, se crea un pipeline que incluye las etapas `assembler`, `scaler` y `pca_initial`. El modelo se ajusta a los datos utilizando `pipeline.fit(X_spark)` y se transforman los datos originales en un nuevo DataFrame con las componentes principales calculadas.

```python
# Seleccionar el número de componentes
explained_var = model.stages[-1].explainedVariance.cumsum()
k = len([var for var in explained_var if var < 0.95]) + 1
```

En este paso, se calcula el número óptimo de componentes principales a retener. Se obtienen las varianzas explicadas acumulativas de las componentes principales calculadas en el paso anterior. Luego, se busca el número mínimo de componentes principales necesarias para explicar al menos el 95% de la varianza total y se almacena en la variable `k`.

```python
# Aplicar PCA con el número óptimo de componentes
pca = PCA(inputCol="scaledFeatures", outputCol="pcaFeatures", k=k)
pipeline = Pipeline(stages=[assembler, scaler, pca])

pca_result = model.transform(X_spark)
```

En este paso, se realiza una segunda ejecución de PCA, pero esta vez con el número óptimo de componentes determinado anteriormente. Se crea una nueva instancia de `PCA` con `k` componentes y se construye un nuevo pipeline.
````python
from pyspark.sql import SparkSession
from ucimlrepo import fetch_ucirepo

# Inicializa la sesión de Spark
spark = SparkSession.builder.appName("CDC Diabetes Health Indicators").getOrCreate()

# Obtiene el dataset
cdc_diabetes_health_indicators = fetch_ucirepo(id=891)

# Obtiene los DataFrames de Pandas
X_pandas = cdc_diabetes_health_indicators.data.features
y_pandas = cdc_diabetes_health_indicators.data.targets

# Convierte los DataFrames de Pandas a DataFrames de PySpark
X_spark = spark.createDataFrame(X_pandas)
y_spark = spark.createDataFrame(y_pandas)

# Muestra las primeras filas para verificar
X_spark.show()
y_spark.show()
````

````python
from pyspark.ml.feature import VectorAssembler, StandardScaler, PCA
from pyspark.ml import Pipeline

# Paso 1: Preparar los datos para PCA
assembler = VectorAssembler(inputCols=X_spark.columns, outputCol="features")
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withStd=True, withMean=False)

# Paso 2: Aplicar PCA
pca_initial = PCA(inputCol="scaledFeatures", outputCol="pcaFeatures", k=len(X_spark.columns))
pipeline = Pipeline(stages=[assembler, scaler, pca_initial])
model = pipeline.fit(X_spark)
transformed = model.transform(X_spark)

# Seleccionar el número de componentes
explained_var = model.stages[-1].explainedVariance.cumsum()
k = len([var for var in explained_var if var < 0.95]) + 1

# Aplicar PCA con el número óptimo de componentes
pca = PCA(inputCol="scaledFeatures", outputCol="pcaFeatures", k=k)
pipeline = Pipeline(stages=[assembler, scaler, pca])
model = pipeline.fit(X_spark)
pca_result = model.transform(X_spark)

# Paso 3: Calcular el tamaño en memoria antes y después de PCA
size_before = len(X_spark.columns) * X_spark.count()  # Tamaño antes de PCA: número de características por número de filas
size_after = k * pca_result.count()  # Tamaño después de PCA: número de componentes PCA por número de filas

# Calcular la reducción
reduction_factor = size_before / size_after

print(f"Tamaño antes de PCA: {size_before}")
print(f"Tamaño después de PCA: {size_after}")
print(f"Factor de reducción: {reduction_factor}")
````


````python
from pyspark.sql import SparkSession
from pyspark.sql.types import IntegerType
from pyspark.sql.functions import monotonically_increasing_id
from pyspark.ml.feature import VectorAssembler, StandardScaler, PCA
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml import Pipeline
from ucimlrepo import fetch_ucirepo

# Inicializa la sesión de Spark
spark = SparkSession.builder.appName("Covertype Dataset Analysis").getOrCreate()

# Obtiene el dataset "covertype"
covertype = fetch_ucirepo(id=31)

# Obtiene los DataFrames de Pandas
X_pandas = covertype.data.features
y_pandas = covertype.data.targets

# Separar la columna objetivo "Cover_Type" del DataFrame de características X_pandas
y_pandas = X_pandas.pop('Cover_Type')

# Convierte los DataFrames de Pandas a DataFrames de PySpark
X_spark = spark.createDataFrame(X_pandas)
y_spark = spark.createDataFrame(y_pandas, IntegerType())

# Añadir una columna de índice a ambos DataFrames
X_spark = X_spark.withColumn("index", monotonically_increasing_id())
y_spark = y_spark.withColumn("index", monotonically_increasing_id())

# Paso 1: Preparar los datos para PCA
assembler = VectorAssembler(inputCols=X_spark.columns[:-1], outputCol="features")
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withStd=True, withMean=False)

# Paso 2: Aplicar PCA
pca = PCA(inputCol="scaledFeatures", outputCol="pcaFeatures", k=20)  # Ajusta el valor de k según sea necesario
pipeline = Pipeline(stages=[assembler, scaler, pca])
model = pipeline.fit(X_spark)
pca_result = model.transform(X_spark)

# Unir los datos de características después de PCA con y_spark
data = pca_result.select("pcaFeatures", "index").join(y_spark, "index", "inner").drop("index")

# Dividir los datos en conjuntos de entrenamiento y prueba
train_ratio = 0.8
test_ratio = 1 - train_ratio
train_data, test_data = data.randomSplit([train_ratio, test_ratio], seed=42)

# Definir el modelo de clasificación (en este caso, Regresión Logística)
classifier = LogisticRegression(featuresCol="pcaFeatures", labelCol="value")

# Crear el pipeline
pipeline = Pipeline(stages=[classifier])

# Entrenar el modelo
model = pipeline.fit(train_data)

# Realizar predicciones en los datos de prueba
predictions = model.transform(test_data)

# Evaluar el rendimiento del modelo
evaluator = MulticlassClassificationEvaluator(labelCol="value", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)

print(f"Accuracy del modelo de clasificación: {accuracy}")


import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import pandas as pd

# Convertir las predicciones a un DataFrame de Pandas
predictions_pandas = predictions.select("prediction", "value").toPandas()

# Generar la matriz de confusión
conf_matrix = confusion_matrix(predictions_pandas['value'], predictions_pandas['prediction'])

# Graficar la matriz de confusión
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Matriz de Confusión')
plt.show()

# Imprimir el reporte de clasificación
report = classification_report(predictions_pandas['value'], predictions_pandas['prediction'])
print("Reporte de Clasificación:\n", report)

# Detener la sesión de Spark
spark.stop()
````
