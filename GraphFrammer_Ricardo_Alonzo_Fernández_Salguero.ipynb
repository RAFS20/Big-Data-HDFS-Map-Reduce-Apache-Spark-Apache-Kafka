# Introducción a GraphFrames en Apache Spark

En el mundo de la ciencia de datos y el análisis de datos a gran escala, Apache Spark se ha convertido en una herramienta fundamental. Spark es un marco de procesamiento distribuido que proporciona una plataforma poderosa para el procesamiento y análisis de datos a gran escala. Dentro del ecosistema de Spark, GraphFrames es una biblioteca especializada que se utiliza para el análisis de datos basado en grafos.

## ¿Qué es un grafo?

Antes de sumergirnos en los detalles de GraphFrames, es importante comprender qué es un grafo. Un grafo es una estructura de datos matemática que se utiliza para representar relaciones entre elementos. Consiste en nodos (vértices) y aristas (bordes) que conectan estos nodos. Los grafos se utilizan en una amplia variedad de aplicaciones, como redes sociales, recomendaciones, análisis de rutas, detección de fraudes y más.

Un grafo se puede representar de dos maneras: grafo dirigido y grafo no dirigido. En un grafo dirigido, las aristas tienen una dirección, lo que significa que la relación entre los nodos va en una dirección específica. En un grafo no dirigido, las aristas no tienen dirección y representan relaciones bidireccionales. Los nodos pueden contener información adicional, conocida como atributos.

## ¿Qué son GraphFrames?

GraphFrames es una biblioteca que extiende las capacidades de procesamiento de datos de Spark para trabajar con datos basados en grafos. Combina la potencia de Spark con la flexibilidad de los grafos para permitir el análisis y la manipulación eficiente de datos conectados. GraphFrames proporciona una interfaz de programación de aplicaciones (API) fácil de usar para trabajar con grafos en Spark.

## Características clave de GraphFrames

Antes de profundizar en la implementación y el uso de GraphFrames, es importante comprender algunas de sus características clave:

### 1. Compatibilidad con DataFrame
GraphFrames está diseñado para ser compatible con las estructuras de datos de DataFrame de Spark. Esto significa que puede combinar fácilmente las operaciones de DataFrame con las operaciones de GraphFrames, lo que hace que trabajar con datos tabulares y datos basados en grafos sea más fluido.

### 2. Soporte para grafos dirigidos y no dirigidos
GraphFrames admite tanto grafos dirigidos como no dirigidos. Puede modelar relaciones unidireccionales y bidireccionales entre los nodos según sus necesidades.

### 3. Operaciones de consulta
GraphFrames proporciona un conjunto de operaciones de consulta similares a SQL que le permiten realizar consultas complejas en los datos del grafo. Puede seleccionar nodos y aristas, aplicar filtros, calcular caminos más cortos, encontrar subgrafos y mucho más.

### 4. Integración con algoritmos de grafos
La biblioteca se integra con algoritmos de grafos comunes, como PageRank, algoritmos de flujo máximo, búsqueda de caminos mínimos y más. Esto facilita la realización de análisis avanzados en los datos del grafo.

### 5. Extensibilidad
GraphFrames es extensible y permite a los usuarios definir sus propias funciones de transformación y análisis para adaptarse a sus necesidades específicas.

```python
def generate_simulated_data(num_users, num_connections):
    # Datos de usuario
    age_groups = {1: range(18, 25), 2: range(25, 40), 3: range(40, 60)}
    interests_by_age_group = {
        1: ['Music', 'Technology'],
        2: ['Sports', 'Art', 'Technology'],
        3: ['Art', 'Music']
    }
    users = pd.DataFrame({
        'id': range(1, num_users + 1),
        'age_group': [random.choice(list(age_groups.keys())) for _ in range(num_users)],
        'location': [random.choice(['City A', 'City B', 'City C']) for _ in range(num_users)]
    })
    users['age'] = users['age_group'].apply(lambda x: random.choice(age_groups[x]))
    users['interests'] = users['age_group'].apply(lambda x: random.choice(interests_by_age_group[x]))
    # Datos de conexiones
    connections = []
    for _ in range(num_connections):
        src = random.randint(1, num_users)
        src_location = users.loc[users['id'] == src, 'location'].values[0]
        if random.random() < 0.90:
            dst_choices = users[users['location'] == src_location]['id'].tolist()
        else:
            dst_choices = users['id'].tolist()
        dst_choices.remove(src)
        dst = random.choice(dst_choices)
        connections.append({'src': src, 'dst': dst})
    connections_df = pd.DataFrame(connections).drop_duplicates()
    return users, connections_df
```

- La función `generate_simulated_data(num_users, num_connections)` genera datos simulados de usuarios y sus conexiones.
- Se definen grupos de edad y los intereses correspondientes a cada grupo.
- Se crea un DataFrame de usuarios con columnas como 'id', 'age_group', 'location', 'age', y 'interests'.
- Luego, se generan conexiones aleatorias entre usuarios, considerando la probabilidad de conexión dentro de la misma ciudad.
- El resultado es un DataFrame de conexiones.

```python
import networkx as nx

def convert_to_networkx(g):
    vertices = g.vertices.toPandas()
    edges = g.edges.toPandas()
    G_nx = nx.Graph()
    for _, row in vertices.iterrows():
        G_nx.add_node(row['id'], **row.to_dict())
    for _, row in edges.iterrows():
        G_nx.add_edge(row['src'], row['dst'])
    return G_nx
```

- La función `convert_to_networkx(g)` toma un objeto GraphFrame `g` y lo convierte en un objeto NetworkX `G_nx`.
- Primero, se obtienen los datos de vértices y aristas como DataFrames de Pandas.
- Luego, se crea un grafo NetworkX (`G_nx`) y se agregan nodos y aristas a partir de los DataFrames.

9. Algoritmos de caminos más cortos (Dijkstra, Bellman-Ford, Floyd-Warshall):

```python
def shortest_path_algorithm(G_nx, algorithm, source_node):
    if algorithm == "dijkstra":
        shortest_paths = nx.single_source_dijkstra_path_length(G_nx, source_node)
    elif algorithm == "bellman-ford":
        shortest_paths, _ = nx.single_source_bellman_ford(G_nx, source_node)
    elif algorithm == "floyd-warshall":
        shortest_paths = dict(nx.all_pairs_shortest_path_length(G_nx))
    else:
        raise ValueError("Algoritmo de camino más corto no válido")
    return shortest_paths
```

- La función `shortest_path_algorithm(G_nx, algorithm, source_node)` calcula los caminos más cortos desde un nodo fuente (`source_node`) utilizando diferentes algoritmos (Dijkstra, Bellman-Ford, Floyd-Warshall).
- Dependiendo del algoritmo especificado, se calculan los caminos más cortos y se almacenan en un diccionario.

```python
def topological_sort_algorithm(G_nx):
    if not nx.is_directed_acyclic_graph(G_nx):
        raise ValueError("El grafo no es un DAG (grafo dirigido acíclico).")
    topological_order = list(topological_sort.topological_sort(G_nx))
    return topological_order
```

- La función `topological_sort_algorithm(G_nx)` calcula el orden topológico del grafo dirigido acíclico (DAG) `G_nx`.

   - Si el grafo es un DAG, se calcula el orden topológico y se visualiza.

El código en su conjunto simula la creación de un grafo de usuarios y conexiones, realiza análisis sobre ese grafo y visualiza diferentes aspectos del mismo. Es una implementación completa que muestra cómo utilizar PySpark y NetworkX para trabajar con datos en forma de grafo y realizar análisis de grafos.

````python
from pyspark.sql import SparkSession
from graphframes import GraphFrame
import pandas as pd
import random
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from networkx.algorithms import community, shortest_paths, minimum_spanning_tree, traversal, topological_sort

# Crear sesión de Spark
def create_spark_session():
    spark = SparkSession.builder \
        .master("local") \
        .appName("GraphXExample") \
        .getOrCreate()
    return spark

# Generar datos simulados con tendencias realistas
def generate_simulated_data(num_users, num_connections):
    # Datos de usuario
    age_groups = {1: range(18, 25), 2: range(25, 40), 3: range(40, 60)}
    interests_by_age_group = {
        1: ['Music', 'Technology'],
        2: ['Sports', 'Art', 'Technology'],
        3: ['Art', 'Music']
    }

    users = pd.DataFrame({
        'id': range(1, num_users + 1),
        'age_group': [random.choice(list(age_groups.keys())) for _ in range(num_users)],
        'location': [random.choice(['City A', 'City B', 'City C']) for _ in range(num_users)]
    })

    users['age'] = users['age_group'].apply(lambda x: random.choice(age_groups[x]))
    users['interests'] = users['age_group'].apply(lambda x: random.choice(interests_by_age_group[x]))

    # Datos de conexiones
    connections = []
    for _ in range(num_connections):
        src = random.randint(1, num_users)
        src_location = users.loc[users['id'] == src, 'location'].values[0]

        # Mayor probabilidad de conexión dentro de la misma ciudad
        if random.random() < 0.90:
            dst_choices = users[users['location'] == src_location]['id'].tolist()
        else:
            dst_choices = users['id'].tolist()

        dst_choices.remove(src)
        dst = random.choice(dst_choices)
        connections.append({'src': src, 'dst': dst})

    connections_df = pd.DataFrame(connections).drop_duplicates()
    return users, connections_df

# Crear GraphFrames
def create_graphframes(spark, users, connections):
    vertices = spark.createDataFrame(users)
    edges = spark.createDataFrame(connections)
    return GraphFrame(vertices, edges)

import networkx as nx

# Convertir a NetworkX
def convert_to_networkx(g):
    vertices = g.vertices.toPandas()
    edges = g.edges.toPandas()

    G_nx = nx.Graph()

    for _, row in vertices.iterrows():
        G_nx.add_node(row['id'], **row.to_dict())

    for _, row in edges.iterrows():
        G_nx.add_edge(row['src'], row['dst'])

    return G_nx

# Detección de comunidades
def detect_communities(G_nx):
    communities = community.greedy_modularity_communities(G_nx)
    return communities

# Visualización del grafo con comunidades
def visualize_graph_with_communities(G_nx, communities):
    plt.figure(figsize=(12, 12))
    pos = nx.kamada_kawai_layout(G_nx)  # Cambio de layout para una mejor visualización

    community_colors = {node: idx for idx, community in enumerate(communities) for node in community}
    colors = [community_colors[n] for n in G_nx.nodes]

    cmap = plt.cm.jet

    nx.draw_networkx_nodes(G_nx, pos, node_color=colors, cmap=cmap, alpha=0.7)
    nx.draw_networkx_edges(G_nx, pos, alpha=0.4)
    nx.draw_networkx_labels(G_nx, pos, font_size=8)

    plt.title("Visualización del Grafo de Usuarios y Conexiones")
    plt.colorbar(plt.cm.ScalarMappable(norm=mcolors.Normalize(vmin=0, vmax=len(communities)-1), cmap=cmap))
    plt.axis('off')
    plt.show()

# Cálculo de Métricas del Grafo
def calculate_graph_metrics(G_nx):
    print("Densidad del Grafo:", nx.density(G_nx))
    print("Centralidad de Intermediación:", nx.betweenness_centrality(G_nx))
    # Agrega aquí más métricas según sea necesario

# Algoritmos de Caminos Más Cortos (Dijkstra, Bellman-Ford, Floyd-Warshall)
def shortest_path_algorithm(G_nx, algorithm, source_node):
    if algorithm == "dijkstra":
        shortest_paths = nx.single_source_dijkstra_path_length(G_nx, source_node)
    elif algorithm == "bellman-ford":
        shortest_paths, _ = nx.single_source_bellman_ford(G_nx, source_node)
    elif algorithm == "floyd-warshall":
        shortest_paths = dict(nx.all_pairs_shortest_path_length(G_nx))
    else:
        raise ValueError("Algoritmo de camino más corto no válido")
    return shortest_paths

# Visualización de Algoritmos de Caminos Más Cortos
def visualize_shortest_paths(G_nx, shortest_paths, source_node, algorithm):
    plt.figure(figsize=(12, 12))
    pos = nx.kamada_kawai_layout(G_nx)

    node_colors = ['limegreen' if node == source_node else 'skyblue' for node in G_nx.nodes()]
    edge_colors = ['red' if (source_node, target) in shortest_paths else 'gray' for target in G_nx.nodes()]

    nx.draw(G_nx, pos, with_labels=True, node_color=node_colors, edge_color=edge_colors, alpha=0.7)
    nx.draw_networkx_labels(G_nx, pos, font_size=8)
    plt.title(f"Visualización del Algoritmo de Camino Más Corto ({algorithm.capitalize()})")
    plt.show()

# Algoritmos de Árbol de Expansión Mínima (Kruskal, Prim)
def minimum_spanning_tree_algorithm(G_nx, algorithm):
    if algorithm == "kruskal":
        mst = minimum_spanning_tree(G_nx)
    elif algorithm == "prim":
        mst = minimum_spanning_tree(G_nx, algorithm='prim')
    else:
        raise ValueError("Algoritmo de árbol de expansión mínima no válido")
    return mst

# Visualización de Algoritmos de Árbol de Expansión Mínima
def visualize_minimum_spanning_tree(G_nx, mst, algorithm):
    plt.figure(figsize=(12, 12))
    pos = nx.kamada_kawai_layout(G_nx)

    nx.draw(G_nx, pos, with_labels=True, node_color='skyblue', edge_color='gray')
    nx.draw_networkx_edges(mst, pos, edge_color='red', width=2)
    plt.title(f"Árbol de Expansión Mínima ({algorithm.capitalize()})")
    plt.show()

# Algoritmos de Búsqueda (DFS, BFS)
def search_algorithm(G_nx, algorithm, start_node):
    if algorithm == "dfs":
        search_tree = traversal.dfs_tree(G_nx, source=start_node)
    elif algorithm == "bfs":
        search_tree = traversal.bfs_tree(G_nx, source=start_node)
    else:
        raise ValueError("Algoritmo de búsqueda no válido")
    return search_tree

# Visualización de Algoritmos de Búsqueda
def visualize_search_tree(G_nx, search_tree, start_node, algorithm):
    plt.figure(figsize=(12, 12))
    pos = nx.kamada_kawai_layout(G_nx)

    nx.draw(G_nx, pos, with_labels=True, node_color='skyblue', edge_color='gray')
    nx.draw_networkx_edges(search_tree, pos, edge_color='red', width=2)
    nx.draw_networkx_nodes(G_nx, pos, nodelist=[start_node], node_color='limegreen')
    plt.title(f"Árbol de Búsqueda ({algorithm.capitalize()})")
    plt.show()

# Orden Topológico
def topological_sort_algorithm(G_nx):
    if not nx.is_directed_acyclic_graph(G_nx):
        raise ValueError("El grafo no es un DAG (grafo dirigido acíclico).")
    topological_order = list(topological_sort.topological_sort(G_nx))
    return topological_order

# Visualización de Orden Topológico
def visualize_topological_sort(G_nx, topological_order):
    plt.figure(figsize=(12, 6))
    pos = nx.kamada_kawai_layout(G_nx)

    node_colors = ['limegreen' if node in topological_order else 'skyblue' for node in G_nx.nodes()]
    nx.draw(G_nx, pos, with_labels=True, node_color=node_colors)
    plt.title("Orden Topológico del Grafo")
    plt.show()

# Visualizaciones Adicionales
def visualize_distributions(users):
    # Distribución de Edad
    plt.figure()
    plt.hist(users['age'], bins=15, color='skyblue', edgecolor='black')
    plt.xlabel('Edad')
    plt.ylabel('Cantidad')
    plt.title('Distribución de Edades de los Usuarios')
    plt.show()

    # Distribución de Intereses
    plt.figure()
    users['interests'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.xlabel('Intereses')
    plt.ylabel('Cantidad')
    plt.title('Distribución de Intereses de los Usuarios')
    plt.show()

# Ejecución Principal
if __name__ == "__main__":
    spark = create_spark_session()
    num_users = 100
    num_connections = 200

    users, connections = generate_simulated_data(num_users, num_connections)
    g = create_graphframes(spark, users, connections)

    G_nx = convert_to_networkx(g)

    communities = detect_communities(G_nx)
    visualize_graph_with_communities(G_nx, communities)

    calculate_graph_metrics(G_nx)
    visualize_distributions(users)

    source_node_dijkstra = 1
    shortest_paths_dijkstra = shortest_path_algorithm(G_nx, "dijkstra", source_node_dijkstra)
    visualize_shortest_paths(G_nx, shortest_paths_dijkstra, source_node_dijkstra, "dijkstra")

    source_node_bellman_ford = 1
    shortest_paths_bellman_ford = shortest_path_algorithm(G_nx, "bellman-ford", source_node_bellman_ford)
    visualize_shortest_paths(G_nx, shortest_paths_bellman_ford, source_node_bellman_ford, "bellman-ford")

    source_node_floyd_warshall = 1
    shortest_paths_floyd_warshall = shortest_path_algorithm(G_nx, "floyd-warshall", source_node_floyd_warshall)

    minimum_spanning_tree_kruskal = minimum_spanning_tree_algorithm(G_nx, "kruskal")
    visualize_minimum_spanning_tree(G_nx, minimum_spanning_tree_kruskal, "kruskal")

    minimum_spanning_tree_prim = minimum_spanning_tree_algorithm(G_nx, "prim")
    visualize_minimum_spanning_tree(G_nx, minimum_spanning_tree_prim, "prim")

    start_node_dfs = 1
    dfs_tree = search_algorithm(G_nx, "dfs", start_node_dfs)
    visualize_search_tree(G_nx, dfs_tree, start_node_dfs, "dfs")

    start_node_bfs = 1
    bfs_tree = search_algorithm(G_nx, "bfs", start_node_bfs)
    visualize_search_tree(G_nx, bfs_tree, start_node_bfs, "bfs")

    if nx.is_directed_acyclic_graph(G_nx):
        topological_order = topological_sort_algorithm(G_nx)
        visualize_topological_sort(G_nx, topological_order)

    # Detener Spark Session al finalizar
    spark.stop()
````
